```python
import numpy as np
import matplotlib.pyplot as plt


def generate_single_mandelbrot(c: complex, max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z**2 + c
    return max_iter


def generate_mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter):
    r1 = np.linspace(xmin, xmax, width)
    r2 = np.linspace(ymin, ymax, height)
    return r1, r2, np.array([[generate_single_mandelbrot(complex(r, i), max_iter) for r in r1] for i in r2])


def draw_mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter):
    d = generate_mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter)
    plt.imshow(d[2], extent=(xmin, xmax, ymin, ymax))
    plt.show()
```

```python
draw_mandelbrot(xmin=-2.0, xmax=1.0, ymin=-1.5, ymax=1.5, width=1000, height=1000, max_iter=256)
```
